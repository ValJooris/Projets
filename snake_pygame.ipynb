```python
import pygame, sys, time, random
from pygame.surfarray import array3d

## Sets up colors for the game using RGB Codes
TURQ = pygame.Color(52, 235, 201)
YELLOW = pygame.Color(255,255,0)
BLACK = pygame.Color(0, 0, 0)
WHITE = pygame.Color(255, 255, 255)
RED = pygame.Color(255, 0, 0)
GREEN = pygame.Color(0, 255, 0) 



class SnakeEnv():
    
    def __init__(self,frame_size_x,frame_size_y):
        '''
        Defines the initial game window size
        '''
        self.frame_size_x = frame_size_x
        self.frame_size_y = frame_size_y
        self.game_window = pygame.display.set_mode((self.frame_size_x, self.frame_size_y))
        self.reset()
    
    def reset(self):
        '''
        Resets the game, along with the default snake size and spawning food.
        '''
        self.game_window.fill(BLACK)
        self.snake_pos = [100, 50]
        self.snake_body = [[100, 50], [100-10, 50], [100-(2*10), 50]]
        self.food_pos = self.spawn_food()
        self.food_spawn = True

        self.direction = "RIGHT"
        self.action = self.direction
        self.score = 0
        self.steps = 0
        print("Game Reset.")
        
    def change_direction(self, action, direction):
        
        if action == "UP" and direction != "DOWN":
            return "UP"
        elif action == "DOWN" and direction != "UP":
            return "DOWN"
        elif action == "RIGHT" and direction != "LEFT":
            return "RIGHT"
        elif action == "LEFT" and direction != "RIGHT":
            return "LEFT"
        else:
            return direction

    def move(self, direction, snake_pos):
        if direction == "UP":
            snake_pos[1] -= 10
        elif direction == "DOWN":
            snake_pos[1] += 10
        elif direction == "RIGHT":
            snake_pos[0] += 10
        elif direction == "LEFT":
            snake_pos[0] -= 10
        return snake_pos
            
        
        
    
    def eat(self):
        '''
        Returns Boolean indicating if Snake has "eaten" the white food square
        '''
        if ((self.snake_pos[0] >= self.food_pos[0] and self.snake_pos[0] <= self.food_pos[0]+10) or (self.snake_pos[0]+10 >= self.food_pos[0] and self.snake_pos[0]+10 <= self.food_pos[0]+10)) and ((self.snake_pos[1] >= self.food_pos[1] and self.snake_pos[1] <= self.food_pos[1]+10) or (self.snake_pos[1]+10 >= self.food_pos[1] and self.snake_pos[1]+10 <= self.food_pos[1]+10)):
            return True
        else:
            return False
   
 
        
        #TODO
   
    def spawn_food(self):
        '''
        Spawns food in a random location on window size
        '''
        
        #TODO
        food_pos = [random.randrange(1, (self.frame_size_x//10)) * 10 -10,
                    random.randrange(1, (self.frame_size_y//10)) * 10 -10]
        return food_pos
        
    def human_step(self,event):
        '''
        Takes human keyboard event and then returns it as an action string
        '''
        
        action = None
        
        if event.type == pygame.QUIT:
            pygame.quit()
            sys.exit()
        ################################################ 
        ########## CONVERT KEYPRESS TO DIRECTION ###### 
        ############################################## 
        elif event.type == pygame.KEYDOWN:
            
            if event.key == pygame.K_UP:
                print("UP")
                
                
             #TODO: down, right and left
            
            if event.key == pygame.K_DOWN:
                print("DOWN")
            if event.key == pygame.K_RIGHT:
                print("RIGHT")
            if event.key == pygame.K_LEFT:
                print("LEFT")    
           
            
            # Esc -> Create event to quit the game
            if event.key == pygame.K_ESCAPE:
                pygame.event.post(pygame.event.Event(pygame.QUIT))
                
        return action
    
    
    
    def display_score(self,color, font, size):
        '''
        Updates the score in top left
        '''
        score_font = pygame.font.SysFont(font, size)
        score_surface = score_font.render('Score : ' + str(self.score), True, color)
        score_rect = score_surface.get_rect()
        score_rect.midtop = (self.frame_size_x/10, 15)
        self.game_window.blit(score_surface, score_rect)
        
    def game_over(self):
        '''
        Checks if the snake has touched the bounding box or itself. If it happens
        call to end_game()
        '''
        #TODO
        if (
        self.snake_pos[0] >= self.frame_size_x
        or self.snake_pos[0] < 0
        or self.snake_pos[1] >= self.frame_size_y
        or self.snake_pos[1] < 0
        or self.snake_pos in self.snake_body[1:]):
            self.end_game()


    def end_game(self):
        '''
        Show only these two messsages (the rest of the screen must be black) -> 
            - "Game has Ended" in the middle of the screen. 
            - Score result
            - 
        '''
        #TODO
        self.game_window.fill(BLACK)
        game_over_font = pygame.font.SysFont(None, 72)
        game_over_surface = game_over_font.render('Game Over', True, WHITE)
        game_over_rect = game_over_surface.get_rect()
        game_over_rect.midtop = (self.frame_size_x / 2, self.frame_size_y / 4)
        self.game_window.blit(game_over_surface, game_over_rect)
    
        score_font = pygame.font.SysFont(None, 48)
        score_surface = score_font.render('Score: ' + str(self.score), True, WHITE)
        score_rect = score_surface.get_rect()
        score_rect.midtop = (self.frame_size_x / 2, self.frame_size_y / 2)
        self.game_window.blit(score_surface, score_rect)
    
        pygame.display.flip()
    
   
        
        time.sleep(3)
        pygame.quit()
        sys.exit()
```

```python
snake_env = SnakeEnv(600,600)

# This is technically a FPS Refresh rate
# Higher number means faster refresh, thus faster snake movement, meaning harder game play
difficulty = 10


# FPS (frames per second) controller
fps_controller = pygame.time.Clock()

# Checks for errors encountered
check_errors = pygame.init()


# Initialise game window
pygame.display.set_caption('Snake Eater') 



# Main logic
while True:
    
    # Check Input from Human Step 
    for event in pygame.event.get():
        if event.type == pygame.KEYDOWN:
            if event.key == pygame.K_LEFT:
                snake_env.action = 'LEFT'
            elif event.key == pygame.K_RIGHT:
                snake_env.action = 'RIGHT'
            elif event.key == pygame.K_UP:
                snake_env.action = 'UP'
            elif event.key == pygame.K_DOWN:
                snake_env.action = 'DOWN'
       
        
    # Check for Direction change based on action
    snake_env.direction = snake_env.change_direction(snake_env.action,snake_env.direction)
    

    #Update Snake Position
    snake_env.snake_pos = snake_env.move(snake_env.direction,snake_env.snake_pos)
    
  


    # Check to see if we ate some food:
    # TODO: insert a new element on the body list (using the new position of snake).
    #       if snake can eat something in the new position, increase score, and spawn food.
    #       else remove the last piece of snake body.
   
    if snake_env.eat():
        snake_env.score += 1
        snake_env.food_spawn = False
        
    snake_env.snake_body.insert(0, list(snake_env.snake_pos))

    if not snake_env.food_spawn:
        snake_env.food_pos = snake_env.spawn_food()
        snake_env.food_spawn = True
    else:
        snake_env.snake_body.pop()


    # ----- Draw Code-----
    # - We start painting everything black
    snake_env.game_window.fill(BLACK)

    # - TODO: Draw the Snake
    for pos in snake_env.snake_body:
        pygame.draw.rect(snake_env.game_window, YELLOW, pygame.Rect(pos[0], pos[1], 20, 20))
  
    # - TODO: Draw Food
    # ---------------------
    pygame.draw.rect(snake_env.game_window, WHITE, pygame.Rect(snake_env.food_pos[0], snake_env.food_pos[1], 20, 20))

    
    # Check if we lost
    snake_env.game_over()
    


    snake_env.display_score(WHITE, 'consolas', 20)
    # Refresh game screen
    pygame.display.update()
    # Refresh rate
    fps_controller.tick(difficulty)
    img = array3d(snake_env.game_window)
```
